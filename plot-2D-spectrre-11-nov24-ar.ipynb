```python
# Importer les bibliothèques nécessaires
import matplotlib.pyplot as plt

# Configurer les paramètres globaux de police et d'apparence des axes
plt.rcParams["font.family"] = "Arial Black"
plt.rcParams['axes.linewidth'] = 2.5  # Largeur des axes

# Création de la figure
fig, ax = plt.subplots(figsize=(60, 20))

# Tracer les deux courbes
plt.plot(file_std5[58000:61000, 0], file_std5[58000:61000, 3], label='STD5', lw=8, marker='o', ms=20, markevery=100)
plt.plot(file_H_renorm_t5[58000:61000, 0], -file_H_renorm_t5[58000:61000, 3], label='HAG5TTT', lw=8)

# Ajouter des légendes
ax.legend(fontsize=50)

# Configurer les labels des axes et la taille des étiquettes
ax.set_title('$\mathbf{\Delta T =600}$ m=3', fontsize=50)
ax.set_xlabel('Energy(ua)', fontsize=50)  # Titre de l'axe X
ax.set_ylabel('Intensity', fontsize=50)    # Titre de l'axe Y
ax.tick_params(axis='both', which='major', labelsize=50)  # Taille des étiquettes

# Ajouter les barres verticales
# Première barre pour le pic max de la courbe STD
peak_std = file_std5[58000:61000, 0][file_std5[58000:61000, 3].argmax()]
ax.axvline(x=peak_std, color='tab:blue', ls='--', lw=8, label='Pic STD5')

# Deuxième barre pour le pic max de la courbe HAG-TTT (en tenant compte de l'inversion de signe)
peak_hag = file_H_renorm_t5[58000:61000, 0][file_H_renorm_t5[58000:61000, 3].argmax()]
ax.axvline(x=peak_hag, color='tab:orange', ls='--', lw=8, label='Pic HAG5TTT')

# Troisième barre pour la valeur 5.91
ax.axvline(x=5.91, color='tab:green', ls='--', lw=8, label='Pic STD20')

# Ajouter une légende pour les barres
ax.legend(fontsize=50)

# Sauvegarder la figure en PDF
plt.savefig('nc2dt600std5hag5tttzoom.pdf', format='pdf')

# Afficher le graphique
plt.show()
```
